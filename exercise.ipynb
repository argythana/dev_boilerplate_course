```python
import pandas as pd
import plotly.express as px
import streamlit as st
```

```python
df = pd.read_csv("data/run4moreProd.activities.csv")
```

```python
df.head()
```

```python
df["date"] = pd.to_datetime(df["createdAt"]).dt.date
```

```python
df.head()
```

```python
df = df.drop("_id", axis=1)
```

```python
date = df.pop("date")
```

```python
df.insert(loc=0, column="date", value=date)
```

```python
df.head()
```

```python
df_by_date = df.groupby("date").sum()#.reset_index()
df_by_date.head(2)
```

```python
df_acts = df.groupby("date").size().to_frame(name="activities").reset_index()
df_acts.head()
```

```python
fig = px.scatter(df, x='distance', y='date', title='Distance by date')

# Show the plot
fig.show()
```

```python
fig = px.scatter(df_acts, x='date', y='activities', title='Activities by date')
fig.show()
```

```python
st.title('Streamlit app')
```

```python
# Scatter plot
st.write("Scatter Plot: Distance by Date")
fig = px.scatter(df, x='date', y='distance', title='Distance by Date')
st.plotly_chart(fig)

# Scatter plot for activities
st.write("Scatter Plot: Activities by Date")
fig_acts = px.scatter(df_acts, x='date', y='activities', title='Activities by Date')
st.plotly_chart(fig_acts)
```
